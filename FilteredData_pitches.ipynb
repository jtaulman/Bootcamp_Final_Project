```python
import pandas as pd
```

```python
df = pd.read_csv('pitches.csv', usecols=['ab_id','pitch_type','code','type','b_count','s_count','outs','on_1b','on_2b','on_3b','px','pz','zone','start_speed','end_speed'])
df
```

```python
df.count()
```

```python
pitches_df = df.dropna()
```

```python
pitches_df.count()
```

```python
pitches_df = pitches_df[['ab_id','pitch_type','code','type','b_count','s_count','outs','on_1b','on_2b','on_3b','px','pz','zone','start_speed','end_speed']]
```

```python
pitches_df.dtypes
```

```python
pitches_type_df = pd.get_dummies(pitches_df[['pitch_type']]) 
```

```python
pitches_type_df.head(15)
```

```python
baseball_df = pd.merge(pitches_type_df, pitches_df, left_index = True, right_index = True)
baseball_df.head()
```

```python
baseball_df = baseball_df[['ab_id','pitch_type_AB', 'pitch_type_CH', 'pitch_type_CU', 'pitch_type_EP', 'pitch_type_FA', 'pitch_type_FC', 'pitch_type_FF', 'pitch_type_FO', 'pitch_type_FS', 'pitch_type_FT', 'pitch_type_IN', 'pitch_type_KC', 'pitch_type_KN', 'pitch_type_PO', 'pitch_type_SC', 'pitch_type_SI', 'pitch_type_SL', 'pitch_type_UN', 'pitch_type','code','type','b_count','s_count','outs','on_1b','on_2b','on_3b','px','pz','zone','start_speed','end_speed']]
baseball_df.head()
```
